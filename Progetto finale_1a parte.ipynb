```python
#Caricare il dataset scaricato dal link https://dataunodc.un.org/dp-crime-corruption-offences

import pandas as pd

file_path = r"C:\Users\marco\OneDrive\Desktop\Progetto Finale\Dataset crimini.xlsx"
data = pd.read_excel(file_path)
data.head()
```

```python
#Filtrare i dati per l'Italia e per unità di misura: Counts
filtered_data = data.loc[
    (data['Country'].isin(["Italy"])) &
    (data['Unit of measurement'] == "Counts")
]
filtered_data.head()
```

```python
#Convertire colonne nel datatype più adatto
filtered_data["Year"] = filtered_data["Year"].astype(int)
filtered_data["VALUE"] = filtered_data["VALUE"].astype(int)
```

```python
#Filtrare i dati per il periodo in cui sono presenti dati relativi ai reati informatici (N.B. in Tableau poi ristretto a 2015-2021 per la gestione di valori null per gli anni 2014 e 2022)
filtered_data = filtered_data[(filtered_data["Year"] >= 2014) & (filtered_data["Year"] <= 2022)]
filtered_data.head()
```

```python
#Trovare l'indice della riga contentente la Category Corruption e anno 2016. Il valore corrispondente nella colonna VALUE è un outlier

print(filtered_data[(filtered_data['Category'] == 'Corruption') & (filtered_data['Year'] == 2016)].index)
```

```python
#Creare un dataframe con solo la categoria Corruption per tutti gli anni tranne il 2016
df_corruption = filtered_data[(filtered_data['Category'] == 'Corruption') & (filtered_data['Year'] != 2016)]

#Calcolare e stampare la media dei valori relativi a df_corruption
media = df_corruption['VALUE'].mean()
media_arrotondata = round(media)
print(media_arrotondata)
```

```python
#Sostituire l'outlier con la media_arrotondata trovata
filtered_data.loc[146, 'VALUE'] = media_arrotondata
```

```python
#Convertire di nuovo la colonna 'VALUE' in int
filtered_data["VALUE"] = filtered_data["VALUE"].astype(int)

#Controllare l'avvenuta sostituzione
filtered_data.loc[146]
```

```python
filtered_data.count()
```

```python
#Eliminare colonne non utili all'analisi
filtered_data = filtered_data.drop(columns=["Iso3_code", "Region", "Subregion", "Indicator", "Dimension", "Sex", "Age", "Source", "Unit of measurement"])

filtered_data.head()
```

```python
filtered_data.info()
```

```python
filtered_data = filtered_data.dropna()
filtered_data.tail()
```

```python
#Caricare il dataset contenente l'indice R&D
file_path = r"C:\Users\marco\OneDrive\Desktop\Progetto Finale\CSV\R&D_final.csv"
df2 = pd.read_csv(file_path, sep=";")
df2.head()
```

```python
df2.rename(columns={"Economy Name": "Country", "Anno": "Year"}, inplace=True)

filtered_data1 = filtered_data.merge(df2, on=['Country', 'Year'], how="left")


filtered_data1.head()
```

```python
filtered_data1 = filtered_data1.drop(columns=["Unnamed: 0"])
```

```python
filtered_data1.head()
```

```python
filtered_data1.rename(columns={"VALUE": "Counts", "value":"R&D"}, inplace=True)
filtered_data1.info()
```

```python
filtered_data_pivot = filtered_data1.pivot(index=['Country', 'Year', "R&D"], columns='Category', values='Counts')
filtered_data_pivot.reset_index(inplace=True)

filtered_data_pivot.head()
```

```python
filtered_data_pivot["Acts against public safety and state security"] = filtered_data_pivot["Unlawful access to a computer system"] + filtered_data_pivot["Unlawful interception or access of computer data"]+ filtered_data_pivot["Unlawful interference with a computer system or computer data"]
filtered_data_pivot.head()
```
